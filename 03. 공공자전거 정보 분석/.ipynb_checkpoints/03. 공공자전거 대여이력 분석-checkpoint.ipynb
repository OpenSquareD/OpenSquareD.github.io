```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
import glob
os.listdir('./data/') # 파일 조회
```

```python
csv_files = glob.glob('./data/서울특별시 공공자전거 대여정보_201906*.csv')
csv_files
```

```python
rent_2019 = pd.DataFrame()
for file in csv_files:
    temp = pd.read_csv(file, encoding='cp949')
    rent_2019 = pd.concat([rent_2019, temp])
rent_2019.head()
```

```python
rent_2020 = pd.read_csv('./data/공공자전거 대여이력 정보_202006.csv', encoding='cp949')
rent_2020.head()
```

```python
bike_rent = pd.concat([rent_2019, rent_2020])
bike_rent
```

```python
bike_station = pd.read_excel('./data/서울특별시 공공자전거 대여소 정보(19.12.9).xlsx')
bike_station
```

```python
# 합계 컬럼 제거
bike_station.drop(1540, inplace=True)
bike_station
```

```python
bike_station.isna().sum()
```

```python
bike_station.columns
```

```python
rent_station = bike_station[['대여소_구', '대여소ID', '대여소주소', '위도', '경도', '거치대수']]
rent_station.columns = ['대여 대여소_구', '대여 대여소번호', '대여 대여소주소', '대여 위도', '대여 경도', '대여 거치대수']

return_station = bike_station[['대여소_구', '대여소ID', '대여소주소', '위도', '경도', '거치대수']]
return_station.columns = ['반납 대여소_구', '반납대여소번호', '반납 대여소주소', '반납 위도', '반납 경도', '반납 거치대수']
```

```python
bike_rent = pd.merge(bike_rent, rent_station, on='대여 대여소번호') 
```

```python
bike_rent = pd.merge(bike_rent, return_station, on='반납대여소번호')
```

```python
bike_rent.head()
```

```python
bike_rent.columns
```

```python
bike_rent['대여일시']
```

```python
bike_rent['반납일시']
```

```python
bike_rent['대여일시'] = pd.to_datetime(bike_rent['대여일시'])
bike_rent['반납일시'] = pd.to_datetime(bike_rent['반납일시'])
```

```python
bike_rent['대여 년도'] = bike_rent['대여일시'].dt.year
bike_rent['대여 일'] = bike_rent['대여일시'].dt.day
bike_rent['대여 요일'] = bike_rent['대여일시'].dt.day_name()
bike_rent['대여 시간'] = bike_rent['대여일시'].dt.hour

bike_rent['반납 년도'] = bike_rent['반납일시'].dt.year
bike_rent['반납 일'] = bike_rent['반납일시'].dt.day
bike_rent['반납 요일'] = bike_rent['반납일시'].dt.day_name()
bike_rent['반납 시간'] = bike_rent['반납일시'].dt.hour
```
