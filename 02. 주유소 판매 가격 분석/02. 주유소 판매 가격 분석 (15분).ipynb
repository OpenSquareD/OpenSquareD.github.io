```python
# 모듈 사용 설정
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./과거_판매가격(주유소)20200726-20200726.csv', encoding='cp949')
df.head()
```

```python
df.info() # 데이터 프레임 정보 조회
```

```python
df['상표'].value_counts() # 브랜드별 확인
```

```python
df['셀프여부'].value_counts() # 셀프 주유소와 일반 주유소 비율 확인
```

```python
df.sort_values('휘발유', ascending=False).head() # 휘발유 판매 가격 상위Top5 조회
```

```python
df.sort_values('휘발유', ascending=True).head() # 휘발유 판매 가격 하위 Top5 조회 
```

```python
df[df['휘발유'] != 0].sort_values('휘발유', ascending=True).head() # 휘발유 판매 가격 하위 Top5 조회 
```

```python
df['시도명'] = df['지역'].str.split().str[0]
df['시군구명'] = df['지역'].str.split().str[1]
df
```

```python
df.groupby('시도명')[['상호']].count() # 시도별 주유소 수 조회
```

```python
df.groupby('시도명')[['휘발유']].mean() # 시도별 평균 휘발유 가격 조회
```

```python
df.groupby(['시도명', '시군구명'])[['휘발유']].mean() # 시도별 평균 휘발유 가격 조회
```

```python
df.groupby(['상표'])[['휘발유']].mean() # 상표별 평균 휘발유 가격 조회
```

```python
df.groupby(['상표', '셀프여부'])[['휘발유']].mean()
```

```python
# 서울 정보 분석
df[df['시도명'] == '서울'].groupby('시군구명')[['휘발유']].mean()
```

```python
df[df['시도명'] == '서울'].groupby(['시군구명', '상표'])[['휘발유']].mean()
```

```python
df.groupby(['시도명', '상표'])[['휘발유']].mean()
```

```python
sns.boxplot(data=df, x='상표', y='휘발유')
```

```python
df_drop_zero = df[df['휘발유'] != 0]
sns.boxplot(data=df_drop_zero, x='상표', y='휘발유')
```

```python
sns.boxplot(data=df_drop_zero, hue='셀프여부', x='상표', y='휘발유')
```

```python
sns.boxplot(data=df_drop_zero, hue='셀프여부', x='시도명', y='휘발유')
```
